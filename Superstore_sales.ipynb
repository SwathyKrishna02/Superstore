Scatter Plot

```python
data = {'sales': [150, 200, 120],
        'another_variable': [45, 60, 35]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
plt.scatter(df['another_variable'], df['sales'], color='blue', alpha=0.7)
plt.title('Scatter Plot of Sales vs Another Variable')
plt.xlabel('Another Variable')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

category-wise sales trend

```python
data = {
    'Date': ['2023-01-01', '2023-02-01', '2023-03-01', '2023-04-01'],
    'CategoryA': [100, 120, 90, 110],
    'CategoryB': [80, 85, 95, 100],
    'CategoryC': [70, 60, 80, 75]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
plt.figure(figsize=(10, 8))
for column in df.columns:
    plt.plot(df.index, df[column], label=column)

plt.title('Sales Trends Across Diff: Categories')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

Discount Impact on Sales

```python
np.random.seed(40)
discount = np.arange(0, 50, 5)
sales = np.random.randint(1200, 15000, len(discount))

data = pd.DataFrame({'Discount': discount, 'Sales': sales})
print(data)
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(x='Discount', y='Sales', data=data, marker='o', color='b')
plt.title('Impact of Discounts on Sales ')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Discount', y='Sales', data=data)
plt.title('Impact of Discounts on Sales')
plt.xlabel('Discount')
plt.ylabel('Sales')
plt.show()
```

Product-wise sales analysis

```python
data = {'Product': ['A', 'B', 'C', 'C', 'D', 'D', 'A', 'B'],
        'Category': ['Clothing', 'Electronics', 'Electronics', 'Clothing','Electronics', 'Electronics', 'Clothing', 'Clothing'],
        'Sales': [1000, 2550, 1800, 2200, 7200, 9000, 2260, 1110]}

data = pd.DataFrame(data)
plt.figure(figsize=(10, 8))
data_grouped = data.groupby('Category')['Sales'].sum().reset_index()
plt.bar(data_grouped['Category'], data_grouped['Sales'], color='skyblue')
plt.title('Total Sales by Category (Bar Chart)')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
data['Date'] = pd.to_datetime(['2024-01-01', '2024-02-01'] * 4, format='%Y-%m-%d')
data_time = data.groupby(['Date', 'Category'])['Sales'].sum().reset_index()

plt.figure(figsize=(10, 8))
for category in data_time['Category'].unique():
    plt.plot(data_time[data_time['Category'] == category]['Date'], data_time[data_time['Category'] == category]['Sales'], marker='o', label=category)

plt.title('Sales Trend Over Time by Category (Line Chart)')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.legend()
plt.show()
```

Region-wise product sales

```python
regions = ['North', 'South', 'East', 'West']
products = ['Product A', 'Product B', 'Product C']
sales = np.array([
    [800, 1100, 750],
    [1000, 1200, 800],
    [950, 1300, 800],
    [100, 200, 400],

])

bar_width = 0.3
r1 = np.arange(len(regions))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.bar(r1, sales[:, 0], width=bar_width, label='Product A')
plt.bar(r2, sales[:, 1], width=bar_width, label='Product B')
plt.bar(r3, sales[:, 2], width=bar_width, label='Product C')

plt.xlabel('Region')
plt.ylabel('Sales')
plt.xticks([r + bar_width for r in range(len(regions))], regions)
plt.legend()

plt.show()
```
